```python
import pandas as pd

# Caminho do arquivo
caminho = '/content/drive/MyDrive/Vai na Web/Copia-de-elegibilidade_credito.csv'

# Lista para guardar os dados
dados = []

# Lê o arquivo linha a linha
with open(caminho, 'r', encoding='utf-8') as f:
    linhas = f.readlines()

# A primeira linha contém os nomes das colunas (vamos ignorar por agora)
for linha in linhas[1:]:
    # Remove aspas, quebra de linha e divide por múltiplos espaços
    partes = linha.replace('"', '').strip().split()

    # Ignora linhas que não têm exatamente 6 colunas
    if len(partes) == 6:
        dados.append(partes)

# Cria o DataFrame
credito = pd.DataFrame(dados, columns=['salario_anual', 'total_dividas', 'historico_pagamento', 'idade', 'credito_solicitado', 'elegibilidade'])

# Converte os dados para tipo numérico, trocando vírgula por ponto
credito['historico_pagamento'] = credito['historico_pagamento'].str.replace(',', '.').astype(float)
for col in ['salario_anual', 'total_dividas', 'idade', 'credito_solicitado', 'elegibilidade']:
    credito[col] = credito[col].astype(int)

# Pronto!
credito.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Seleciona as variáveis independentes e o alvo
X = credito[['salario_anual', 'total_dividas', 'historico_pagamento', 'idade', 'credito_solicitado']]
y = credito['elegibilidade']

# Divide em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

# Cria e treina o modelo
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Faz previsões
y_pred = knn.predict(X_test)

# Avalia o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia do modelo KNN: {accuracy:.2f}")
```
